Add Trigger Word Command

```python
import os

def append_to_files(directory, prefix):
    for filename in os.listdir(directory):
        if filename.endswith(".txt"):
            file_path = os.path.join(directory, filename)
            with open(file_path, 'r+') as file:
                content = file.read()
                file.seek(0, 0)
                file.write(prefix + content)

# Example usage
directory_path = "/content/DeepDanbooru/sibear/dataset"
text_to_append = "sibear, sidebangs hairstyle with earring,"

append_to_files(directory_path, text_to_append)
```

Delete Text File Command (If There Mistake)

```python
import os

def delete_txt_files(directory):
    for filename in os.listdir(directory):
        if filename.endswith(".txt"):
            file_path = os.path.join(directory, filename)
            os.remove(file_path)

# Example usage
directory_path = "/content/DeepDanbooru/sibear/dataset"

delete_txt_files(directory_path)
```

```python
import os
import zipfile
from tqdm import tqdm

def zip_directory(directory_path, zip_file_name):
    with zipfile.ZipFile(zip_file_name, 'w') as zip_file:
        file_paths = []
        for root, dirs, files in os.walk(directory_path):
            file_paths.extend([os.path.join(root, file) for file in files])

        for file_path in tqdm(file_paths, desc="Zipping", unit="file"):
            zip_file.write(file_path, os.path.relpath(file_path, directory_path))

# Example usage
directory_path = "/content/DeepDanbooru/sibear/dataset"
zip_file_name = "/content/sibear-danbooru-tag/sibear-danbooru-tag.zip"

zip_directory(directory_path, zip_file_name)
```

Upload Tag For Diffusers

```python
import os
import csv

def list_files(directory):
    file_list = []
    for root, dirs, files in os.walk(directory):
        for file in files:
            file_list.append(os.path.join(root, file))
    return file_list

def read_txt_content(txt_file):
    with open(txt_file, 'r') as f:
        content = f.read()
    return content

def create_csv(file_data, csv_filename):
    with open(csv_filename, 'w', newline='') as csvfile:
        fieldnames = ['file_name', 'additional_feature']
        writer = csv.DictWriter(csvfile, fieldnames=fieldnames)

        writer.writeheader()

        for data in file_data:
            image_file = f"{data['number']}.jpg"
            additional_feature = data['content']

            writer.writerow({'file_name': image_file, 'additional_feature': additional_feature})

def main(directory, csv_filename):
    files = list_files(directory)
    file_data = []

    for file_path in files:
        file_name = os.path.basename(file_path)

        if file_name.endswith(".txt"):
            file_number = int(file_name.split('.')[0])
            txt_content = read_txt_content(file_path)
            file_data.append({'number': file_number, 'content': txt_content})

    sorted_data = sorted(file_data, key=lambda x: x['number'])
    create_csv(sorted_data, csv_filename)

if __name__ == "__main__":
    directory_path = "/content/DeepDanbooru/sibear/dataset"  # replace with the actual directory path
    csv_file_path = "/content/DeepDanbooru/sibear/dataset/metadata.csv"  # replace with the desired CSV file name

    main(directory_path, csv_file_path)
```
